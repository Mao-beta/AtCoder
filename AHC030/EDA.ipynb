```python
import random
import matplotlib.pyplot as plt
import seaborn as sns


# 値が1以上だったときに本当にその範囲に油田が存在すると考えてよい最大のkは？
# 

eps = 0.01
for k in range(1, 17):
    for v in range(1, 17):
        mu = (k-v) * eps + v * (1-eps)
        var = k * eps * (1-eps)
        print(eps, k, v, round(mu, 2), round(var, 2), sep="\t\t")
```
